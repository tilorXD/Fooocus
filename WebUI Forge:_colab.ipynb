```python
import os

# ==============================================================================
# üìù TU LISTA DE DESCARGAS (EDITA AQU√ç)
# ==============================================================================
# REGLA DE ORO: ¬°No olvides poner una coma (,) al final de cada l√≠nea!
# Formato: ("URL", "NOMBRE_ARCHIVO", "TIPO")

DOWNLOAD_LIST = [
    # --- CHECKPOINTS (Modelos Principales) ---
    ("https://civitai.com/api/download/models/288982", "Juggernaut_XL_v9.safetensors", "checkpoint"),
    ("https://civitai.com/api/download/models/691639?type=Model&format=SafeTensor&size=pruned&fp=fp32", "flux_dev.safetensors", "checkpoint"),
    ("https://huggingface.co/Kijai/flux-fp8/resolve/main/flux1-dev-fp8.safetensors", "flux1-dev-fp8.safetensors", "checkpoint"),
    # --- LORAS (Tus modelos del error) ---
    # F√≠jate que ahora todas las l√≠neas terminan en coma ,
    ("https://civitai.com/api/download/models/801399?type=Model&format=SafeTensor", "aidmaGothGirl-FLUX-V0.1.safetensors", "lora"),
    ("https://civitai.com/api/download/models/1857758?type=Model&format=SafeTensor", "microbikiniv12_FLUX.safetensors", "lora"),

    # --- Agrega m√°s aqu√≠ abajo, siempre con coma al final ---
    # ("https://...", "mi_modelo.safetensors", "lora"),
]

# ==============================================================================
# ‚öôÔ∏è MOTOR DE DESCARGA (NO TOCAR)
# ==============================================================================
print("üöÄ Iniciando Gestor de Descargas para Fooocus...")

# Rutas base de Fooocus en Colab
base_path = "/content/Fooocus/models"

# Mapeo de carpetas
folders = {
    "checkpoint": f"{base_path}/checkpoints",
    "checkpoints": f"{base_path}/checkpoints",
    "Checkpoints": f"{base_path}/checkpoints",

    "lora": f"{base_path}/loras",
    "loras": f"{base_path}/loras",
    "LORAs": f"{base_path}/loras",

    "vae": f"{base_path}/vae_approx",
}

# Crear carpetas si no existen
for path in folders.values():
    os.makedirs(path, exist_ok=True)

for url, filename, type_key in DOWNLOAD_LIST:
    # Verificar si el tipo es v√°lido
    if type_key not in folders:
        print(f"‚ö†Ô∏è Tipo desconocido: '{type_key}' para {filename}. Saltando...")
        continue

    destination = folders[type_key]
    file_path = os.path.join(destination, filename)

    # Comprobar si ya existe
    if os.path.exists(file_path):
        print(f"‚úÖ Ya existe: {filename}")
    else:
        print(f"‚¨áÔ∏è Descargando: {filename} en [{type_key}]...")
        # Usamos user-agent para evitar bloqueos de CivitAI
        !aria2c --console-log-level=error -c -x 16 -s 16 -k 1M --user-agent="Mozilla/5.0" "{url}" -d "{destination}" -o "{filename}"

print("\n‚ú® ¬°Descargas finalizadas! Ya puedes iniciar Fooocus.")
```
