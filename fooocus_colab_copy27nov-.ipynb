# **Gestor de Descargas (Modo SDXL)**

```python
import os

# ==============================================================================
# üìù LISTA DE DESCARGAS (SOLO MODELOS SDXL)
# ==============================================================================
# Formato: ("URL", "NOMBRE_ARCHIVO.extensi√≥n", "TIPO")
# Tipos: "checkpoint", "lora"

DOWNLOAD_LIST = [
    # --- CHECKPOINT PRINCIPAL (Motor SDXL) ---
    # Juggernaut XL v9 (El rey del fotorealismo en Fooocus)
    ("https://civitai.com/api/download/models/456194?type=Model&format=SafeTensor", "juggernautXL_v9_RunDiffusionPhoto_v2.safetensors", "checkpoint"),
    ("https://civitai.com/api/download/models/1920523?type=Model&format=SafeTensor&size=pruned&fp=fp16", "epicrealismXL_vxviiCrystalclear.safetensors", "checkpoint"),

    ("https://civitai.com/api/download/models/294470?type=Model&format=SafeTensor&size=pruned&fp=fp16", "realisticStockPhoto_v20.safetensors", "checkpoint"),
    ("https://civitai.com/api/download/models/1075446?type=Model&format=SafeTensor&size=pruned&fp=fp16", "realismByStableYogi_v5XLFP16.safetensors", "checkpoint"),


    # --- LORAS (Accesorios SDXL) ---
    # Ejemplo: Detail Tweaker XL (Mejora detalles)

    ("https://civitai.com/api/download/models/135867?type=Model&format=SafeTensor", "Detail_Tweaker_XL.safetensors", "lora"),
    ("https://civitai.com/api/download/models/177674?type=Model&format=SafeTensor", "Nudify.safetensors", "lora"),
    ("https://civitai.com/api/download/models/133037?type=Model&format=SafeTensor", "SDXL_Toples.safetensors", "lora"),
    ("https://civitai.com/api/download/models/288679?type=Model&format=SafeTensor", "iphonemirror.safetensors", "lora"),
    ("https://civitai.com/api/download/models/128461?type=Model&format=SafeTensor","PerfectEyesXL.safetensors","lora"),

    ("https://civitai.com/api/download/models/140555?type=Model&format=SafeTensor", "POVmissionary.safetensors", "lora"),


    # ‚ö†Ô∏è NO AGREGUES LORAS DE FLUX AQU√ç, DAR√ÅN ERROR.
    # Agrega tus LoRAs de SDXL aqu√≠ abajo (recuerda la coma al final):
    # ("LINK...", "nombre.safetensors", "lora"),
]

# ==============================================================================
# ‚öôÔ∏è MOTOR DE DESCARGA
# ==============================================================================
print("üöÄ Iniciando Gestor de Descargas para Fooocus...")

base_path = "/content/Fooocus/models"

# Diccionario de carpetas de Fooocus
folders = {
    "checkpoint": f"{base_path}/checkpoints",
    "checkpoints": f"{base_path}/checkpoints",
    "lora": f"{base_path}/loras",
    "loras": f"{base_path}/loras",
    "vae": f"{base_path}/vae_approx",
}

# Crear carpetas
for path in folders.values():
    os.makedirs(path, exist_ok=True)

for url, filename, type_key in DOWNLOAD_LIST:
    if type_key not in folders:
        print(f"‚ö†Ô∏è Tipo desconocido: '{type_key}'. Saltando...")
        continue

    destination = folders[type_key]
    file_path = os.path.join(destination, filename)

    if os.path.exists(file_path):
        print(f"‚úÖ Ya existe: {filename}")
    else:
        print(f"‚¨áÔ∏è Descargando: {filename}...")
        # Usamos user-agent para evitar bloqueos
        !aria2c --console-log-level=error -c -x 16 -s 16 -k 1M --user-agent="Mozilla/5.0" "{url}" -d "{destination}" -o "{filename}"

print("\n‚ú® ¬°Modelos listos! Ya tienes el motor correcto.")
```
