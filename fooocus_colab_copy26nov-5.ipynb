# **Gestor de Descargas (Modo SDXL)**

```python
import os

# ==============================================================================
# 📝 LISTA DE DESCARGAS (SOLO MODELOS SDXL)
# ==============================================================================
# Formato: ("URL", "NOMBRE_ARCHIVO.extensión", "TIPO")
# Tipos: "checkpoint", "lora"

DOWNLOAD_LIST = [
    # --- CHECKPOINT PRINCIPAL (Motor SDXL) ---
    # Juggernaut XL v9 (El rey del fotorealismo en Fooocus)
    ("https://civitai.com/api/download/models/456194?type=Model&format=SafeTensor", "juggernautXL_v9_RunDiffusionPhoto_v2.safetensors", "checkpoint"),

    # --- LORAS (Accesorios SDXL) ---
    # Ejemplo: Detail Tweaker XL (Mejora detalles)
    ("https://civitai.com/api/download/models/135867?type=Model&format=SafeTensor", "Detail_Tweaker_XL.safetensors", "lora"),

    # ⚠️ NO AGREGUES LORAS DE FLUX AQUÍ, DARÁN ERROR.
    # Agrega tus LoRAs de SDXL aquí abajo (recuerda la coma al final):
    # ("LINK...", "nombre.safetensors", "lora"),
]

# ==============================================================================
# ⚙️ MOTOR DE DESCARGA
# ==============================================================================
print("🚀 Iniciando Gestor de Descargas para Fooocus...")

base_path = "/content/Fooocus/models"

# Diccionario de carpetas de Fooocus
folders = {
    "checkpoint": f"{base_path}/checkpoints",
    "checkpoints": f"{base_path}/checkpoints",
    "lora": f"{base_path}/loras",
    "loras": f"{base_path}/loras",
    "vae": f"{base_path}/vae_approx",
}

# Crear carpetas
for path in folders.values():
    os.makedirs(path, exist_ok=True)

for url, filename, type_key in DOWNLOAD_LIST:
    if type_key not in folders:
        print(f"⚠️ Tipo desconocido: '{type_key}'. Saltando...")
        continue

    destination = folders[type_key]
    file_path = os.path.join(destination, filename)

    if os.path.exists(file_path):
        print(f"✅ Ya existe: {filename}")
    else:
        print(f"⬇️ Descargando: {filename}...")
        # Usamos user-agent para evitar bloqueos
        !aria2c --console-log-level=error -c -x 16 -s 16 -k 1M --user-agent="Mozilla/5.0" "{url}" -d "{destination}" -o "{filename}"

print("\n✨ ¡Modelos listos! Ya tienes el motor correcto.")
```
